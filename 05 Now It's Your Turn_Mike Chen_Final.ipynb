```python
#Import Packages
%matplotlib inline

from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier, MLPRegressor
from dmba import classificationSummary

import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
#Import Data
car_data=pd.read_csv('05_ToyotaCorolla.csv')
```

```python
#Check data
car_data.head();
```

```python
# Get dummy variables
car_data = pd.get_dummies(car_data, columns=['Fuel_Type'], drop_first=True)
```

```python
# Select predictors and target variable
predictors = ['Age_08_04', 'KM', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'HP', 'Automatic', 'Doors',
              'Quarterly_Tax', 'Mfr_Guarantee', 'Guarantee_Period', 'Airco',
              'Automatic_airco', 'CD_Player', 'Powered_Windows', 'Sport_Model', 'Tow_Bar']
target = ['Price']

```

```python
# Separate features X and target y
X = car_data[predictors]
y = car_data[target]
```

```python
y
```

```python
# Split data into train and test sets. Partition data into training (60%) and test (40%) sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
# Standardize features and target using MinMaxScaler
X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

y_train_scaled = y_scaler.fit_transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Build and train the neural network model
model = MLPRegressor(hidden_layer_sizes=(2), activation='relu', solver='adam', random_state=1)
model.fit(X_train_scaled, y_train_scaled.ravel())
```

```python
#MLPRegressor is used because we want to predict the price of a used Toyota Corolla based on its specifications and price is a continuous value
```

```python
# Make predictions on price of Toyota Corolla based on specifications 
y_test_scaled_pred = model.predict(X_test_scaled)
```
