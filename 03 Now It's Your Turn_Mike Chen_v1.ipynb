```python
%matplotlib inline

from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
```

```python
housing_data=pd.read_csv('03_BostonHousing.csv')
#rename Cat.MEDV for easier data handling 
housing_data.rename(columns={'CAT. MEDV': 'CAT_MEDV'}, inplace=True) 
```

```python
housing_data
```

```python
#The goal is to predict the median house price in new tracts based on information such as crime rate, pollution, and number of rooms.
```

```python
# Question 1 Answer = Test sets allow the predictor model to run predictions on unseen data avoiding overfitting and setting a benchmark to evaluate the model, while training sets are used to teach the model. This allows the model to run predictions on seen data. With both training and test sets one is able to get a sense of the prediction model's performance/reliability. 
```

```python
from sklearn.linear_model import LinearRegression

# Select predictors and target variable
X = housing_data[['CRIM', 'CHAS', 'RM']]  # Predictors
y = housing_data['MEDV']  # Target variable

# Fit the multiple Linear Regression model
model = LinearRegression()
model= model.fit(X, y)

# Get coefficients and intercept
intercept = model.intercept_
coefficients = model.coef_
```

```python
# Step 5: Print the regression equation
print(f"Regression Equation: MEDV = {intercept:.2f} + {coefficients[0]:.2f}*CRIM + {coefficients[1]:.2f}*CHAS + {coefficients[2]:.2f}*RM")
```
