```python
#Import Packages
%matplotlib inline

from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
```

```python
#Import Data
household_data=pd.read_csv('BathSoapHousehold.csv')
```

```python
household_data
```

```python
#All predictors written out for ease. 

#Member id

#Demographics = SEC, FEH, MT, SEX, AGE, EDU, HS, CHILD, CS, Affluence Index

#Purchase summary over the period = No. of Brands, Brand Runs, Total Volume, No. of Trans, Value, Trans / Brand Runs, Vol/Tran, Avg. Price

#Purchase within promotion = Pur Vol No Promo - %, Pur Vol Promo 6 %, Pur Vol Other Promo %, 

#Brandwise purchase =  Br. Cd. 57, 144, Br. Cd. 55, Br. Cd. 272,  Br. Cd. 286,  Br. Cd. 24,  Br. Cd. 481, Br. Cd. 352,  Br. Cd. 5 ,  Others 999

#Price categorywise purchase = Pr Cat 1, Pr Cat 2, Pr Cat 3, Pr Cat 4 

#Selling propositionwise purchase = PropCat 5, PropCat 6, PropCat 7, PropCat 8, PropCat 9, PropCat 10, PropCat 11, PropCat 12, PropCat 13, PropCat 14, PropCat 15
```

```python
# Create a subset of features related to purchase behavior (volume, frequency, susceptibility to discounts, and brand loyalty)

purchase_behavior_predictors = household_data[['No. of Brands', 'Brand Runs', 'Total Volume','No. of  Trans', 'Value', 'Trans / Brand Runs', 'Vol/Tran', 'Avg. Price ',  'Pur Vol No Promo - %', 'Pur Vol Promo 6 %', 'Pur Vol Other Promo %', 'Br. Cd. 57, 144', 'Br. Cd. 55', 'Br. Cd. 272', 'Br. Cd. 286', 'Br. Cd. 24', 'Br. Cd. 481', 'Br. Cd. 352', 'Br. Cd. 5', 'Others 999']]

```

```python
purchase_behavior_predictors
```

```python
# Normalize the data
scaler = StandardScaler()
scaled_behavior_predictors = scaler.fit_transform(purchase_behavior_predictors)
```

```python
print(pd.DataFrame(scaled_purchase_behavior_predictors, columns=purchase_behavior_predictors.columns))
```

```python
# Apply k-means clustering

from sklearn.cluster import KMeans

kmeans_purchase_behavior = KMeans(n_clusters=3, random_state=1)  # Choose k based on analysis
kmeans_purchase_behavior.fit(scaled_purchase_behavior_predictors)
```

```python
# Assign cluster labels to the data
household_data['Purchase_Behavior_Cluster'] = kmeans_behavior.labels_
```
