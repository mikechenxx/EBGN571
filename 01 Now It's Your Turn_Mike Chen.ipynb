```python
#import packages
import numpy as np 

%matplotlib inline

import os 
import calendar 
from pathlib import Path 
import pandas as pd 
import networkx as nx
from pandas.plotting import scatter_matrix, parallel_coordinates 
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pylab as plt

import matplotlib.pyplot as plt 
import plotly
```

```python
#reading provided csv file
ridingmowers_df = pd.read_csv('01_RidingMowers.csv')
```

```python
ridingmowers_df.head(27) 
```

```python
#creating a rough scatter plot
ridingmowers_df.plot.scatter(x='Lot_Size', y='Income', legend = True)
plt.xlabel('Lot_Size')
plt.ylabel('Income')

plt.tight_layout()
plt.show()
```

```python
#making the scatter plot look nice w labels and legend
#making scatter plot color-coded by owner/nonowner



ridingmowers_df['Ownership_Value'] = ridingmowers_df['Ownership'].apply(lambda x: 1 if x== 'Owner' else 0) #make new column to select status of ownership 

ridingmowers_df.plot.scatter(x= 'Lot_Size', y= 'Income', c=['black' if p== 1 else 'red' for p in ridingmowers_df.Ownership_Value], figsize= [7,5]) #plot scatter plot based on status of ownership

plt.legend(handles= [plt.scatter([], [], c='black', label = 'Owner'), plt.scatter([], [], c= 'red', label= 'Nonowner')], loc=4)   #making a legend for graph based on owner/nonowner



plt.xlabel('Lot_Size')
plt.ylabel('Income')

plt.tight_layout
plt.show()
```

```python
#Insight = Owners tend to have larger lot sizes and higher income compared to the nonowners, so we may want to prioritize targeting owners first and then nonowners based on the data. 
```
